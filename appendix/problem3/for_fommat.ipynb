# 更改格式

```python
import os
import pandas as pd

# Set the path where the CSV files are located
path = 'sol3_'

# List all files in the directory
files = [file for file in os.listdir(path) if file.endswith('scheduling_results.csv')]

# Display the list of CSV files found
files
```

```python
# Full path to the files
csv_files_full_path = [os.path.join(path, file) for file in files]

# Read and concatenate all CSV files into a single DataFrame
df_list = [pd.read_csv(file, header=None) for file in csv_files_full_path]
combined_df = pd.concat(df_list, ignore_index=True)

# Since the user has provided the desired format in the second image, let's take a look at the first few rows
# to understand the current format of the data.
combined_df
```

```python
# It seems like the values in '临时工' are strings with a .0 at the end.
# We will remove any potential non-numeric characters, then convert to float and then to int.

# Remove non-numeric characters (except for the decimal point)
combined_df['临时工'] = combined_df['临时工'].replace({r'[^\d.]': ''}, regex=True)

# Convert the column to float first (to handle any remaining decimal points properly)
combined_df['临时工'] = combined_df['临时工'].astype(float)

# Now convert to int
combined_df['临时工'] = combined_df['临时工'].astype(int)

# Check the changes
combined_df.to_csv("结果表5.csv",index=None)
```
